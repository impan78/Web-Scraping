```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_largest_companies_in_the_United_States_by_revenue'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
# print(soup)
```

```python
table = soup.find(name='table')
table
```

```python
titles = table.find_all('th')
titles
```

```python
titles_list = [title.text.strip() for title in titles]
titles_list
```

```python
df = pd.DataFrame(columns=titles_list)
df
```

```python
table.find_all('tr')[1].find_all('td')[1].text.strip()
```

```python
all_rows = table.find_all('tr')

rows = [rows_.find_all('td') for rows_ in all_rows[1:]]

values = [[value.text.strip() for value in row] for row in rows]
values
```

```python
value_df = pd.DataFrame(values)
value_df
```

```python
value_df.columns = df.columns
value_df
```

```python
value_df.to_csv('US_Largest_Companies.csv', index=False)
```
