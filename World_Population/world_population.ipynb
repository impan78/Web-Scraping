```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
URL = "https://www.worldometers.info/world-population/"
```

```python
response = requests.get(URL)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
tables = soup.find_all("table")
len(tables)
```

```python
def tabel_scrap(table):
    return table.find_all("tr")
```

```python
def rows_scrap(rows, scrapping="Columns Name"):
    if scrapping == "Columns Name":
        return [header.text.strip() for header in rows[0].find_all("th")]
    
    data = []
    for row in rows[1:]:
        cells = row.find_all("td")
        data.append([cell.text.strip() for cell in cells])
    return data
```

```python
def create_df(data, columns):
    return pd.DataFrame(data, columns=columns)
```

```python
table0 = tables[0]
table0_rows = tabel_scrap(table0)
```

```python
table0_col_name = rows_scrap(table0_rows)
table0_col_name
```

```python
table0_data = rows_scrap(table0_rows, scrapping="data")
table0_data
```

```python
world_population_history = create_df(table0_data, table0_col_name)
world_population_history
```

```python
table1 = tables[1]
table1_rows = tabel_scrap(table1)
```

```python
table1_col_name = rows_scrap(table1_rows)
table1_col_name
```

```python
table1_data = rows_scrap(table1_rows, scrapping="data")
table1_data
```

```python
world_population_forecast = create_df(table1_data, table1_col_name)
world_population_forecast
```

```python
table2 = tables[2]
table2_rows = tabel_scrap(table2)
```

```python
table2_col_name = rows_scrap(table2_rows)
table2_col_name
```

```python
table2_data = rows_scrap(table2_rows, scrapping="data")
table2_data
```

```python
sumary_table = create_df(table2_data, table2_col_name)
sumary_table
```

```python
table3 = tables[3]
table3_rows = tabel_scrap(table3)
```

```python
table3_col_name = rows_scrap(table3_rows)
table3_col_name
```

```python
table3_data = rows_scrap(table3_rows, scrapping="data")
table3_data
```

```python
world_population_Region = create_df(table3_data, table3_col_name)
world_population_Region.drop('#', inplace=True, axis=1)
world_population_Region
```

```python
table4 = tables[4]
table4_rows = tabel_scrap(table4)
```

```python
table4_col_name = rows_scrap(table4_rows)
table4_col_name
```

```python
table4_data = rows_scrap(table4_rows, scrapping="data")
table4_data[:5]
```

```python
world_population_country = create_df(table4_data, table4_col_name)
world_population_country.drop('#', inplace=True, axis=1)
world_population_country
```

```python
world_population_history.to_csv("extracted_files/world_population_history.csv", index=False)
world_population_forecast.to_csv("extracted_files/world_population_forecast.csv", index=False)
sumary_table.to_csv("extracted_files/sumary_table.csv", index=False)
world_population_Region.to_csv("extracted_files/world_population_Region.csv", index=False)
world_population_country.to_csv("extracted_files/world_population_country.csv", index=False)
```
